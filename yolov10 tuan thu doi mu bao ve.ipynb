**PHÁT HIỆN ĐỐI TƯỢNG VỚI VIDEO**

```python
from ultralytics import YOLOv10
import cv2
import os

# Đường dẫn đến mô hình
MODEL_PATH = "/content/drive/MyDrive/yolov10/yolov10n.pt"

# Tải mô hình
model = YOLOv10(MODEL_PATH)

# Đường dẫn đến video đầu vào
VIDEO_PATH = "/content/drive/MyDrive/yolov10/00_06_online-video-cutter.com.mp4"

# Đường dẫn để lưu video sau khi phát hiện
OUTPUT_DIR = "/content/drive/MyDrive/yolov10/output_videos"
OUTPUT_VIDEO_PATH = os.path.join(OUTPUT_DIR, "video_sau_khi_phat_hien.mp4")

# Tạo thư mục đầu ra nếu chưa tồn tại
if not os.path.exists(OUTPUT_DIR):
    os.makedirs(OUTPUT_DIR)
    print(f"Tạo thư mục: {OUTPUT_DIR}")

# Mở video đầu vào
cap = cv2.VideoCapture(VIDEO_PATH)

# Kiểm tra xem video có mở thành công không
if not cap.isOpened():
    print(f"Không thể mở video: {VIDEO_PATH}")
    exit()

# Lấy thông tin về video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec video
fps = int(cap.get(cv2.CAP_PROP_FPS))      # Số khung hình trên giây
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))  # Chiều rộng khung hình
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))  # Chiều cao khung hình

# Tạo đối tượng VideoWriter để ghi video
out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, fps, (width, height))

# Kiểm tra xem VideoWriter có khởi tạo thành công không
if not out.isOpened():
    print(f"Không thể ghi video vào: {OUTPUT_VIDEO_PATH}")
    cap.release()
    exit()

frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Chạy mô hình trên từng khung hình
    result = model(source=frame)[0]

    # Lấy khung hình đã chú thích
    annotated_frame = result.plot()

    # Ghi khung hình đã chú thích vào video đầu ra
    out.write(annotated_frame)

    frame_count += 1
    if frame_count % 10 == 0:
        print(f"Đã xử lý {frame_count} khung hình")

# Giải phóng bộ nhớ
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Video đã chú thích được lưu tại {OUTPUT_VIDEO_PATH}")
```

**HUẤN LUYỆN MÔ HÌNH**

```python
import os
import torch

# Đặt biến môi trường để quản lý sự phân mảnh bộ nhớ
os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'max_split_size_mb:128'
torch.cuda.empty_cache()

# Cập nhật các tham số
yaml_path = "/content/drive/MyDrive/yolov10/Safety_Helmet_Dataset/data.yaml"
epochs = 50
img_size = 416  # giảm kích thước ảnh thêm
batch_size = 64  # giảm kích thước batch thêm
workers = 1  # sử dụng ít workers hơn

# Bắt đầu huấn luyện với các tham số đã cập nhật
model.train(
    data=yaml_path,
    epochs=epochs,
    batch=batch_size,
    imgsz=img_size,
    workers=workers,  # giảm số lượng workers
    amp=True
)
```
